```python
import pandas as pd
import numpy as np
```

```python
drugcomb=pd.read_csv(r'summary_v_1_5.csv')
```

```python
drugcomb = drugcomb[drugcomb['drug_row'].notnull() & drugcomb['drug_col'].notnull()]
```

```python
drugcomb = drugcomb.drop_duplicates()
```

```python
synergy_cols = ["synergy_zip", "synergy_loewe", "synergy_hsa", "synergy_bliss"]
drugcomb = drugcomb.dropna(subset=synergy_cols, how="all")
```

```python
columns_to_keep = ['drug_row', 'drug_col', 'cell_line_name', 'tissue_name',
                   'synergy_zip', 'synergy_loewe', 'synergy_hsa', 'synergy_bliss',
                   'ic50_row', 'ic50_col', 'ri_row', 'ri_col', 
                   'css_row', 'css_col', 'css_ri',
                   'drug_row_target_name', 'drug_col_target_name']

drugcomb = drugcomb[columns_to_keep]
```

```python
drugcomb['drug_row_target_name'].fillna('Unknown', inplace=True)
drugcomb['drug_col_target_name'].fillna('Unknown', inplace=True)
```

```python
drugcomb['drug_row'] = drugcomb['drug_row'].str.strip().str.upper()
drugcomb['drug_col'] = drugcomb['drug_col'].str.strip().str.upper()
```

```python
num_cols = ['ic50_row','ic50_col','ri_row','ri_col','css_row','css_col','css_ri']
drugcomb[num_cols] = drugcomb[num_cols].fillna(drugcomb[num_cols].median())
```

```python
drugcomb['tissue_name'] = drugcomb['tissue_name'].str.strip().str.upper()
drugcomb['cell_line_name'] = drugcomb['cell_line_name'].str.strip().str.upper()
```

```python
drugcomb = drugcomb.drop_duplicates()
```

```python
drugcomb[['drug_row', 'drug_col']] = np.sort(drugcomb[['drug_row','drug_col']], axis=1)
drugcomb = drugcomb.drop_duplicates()
```

```python
drugcomb = drugcomb.dropna(subset=['synergy_zip'])
```

```python
drugcomb = drugcomb.reset_index(drop=True)
```

```python
drugcomb.to_csv("drugcomb_clean.csv", index=False)
```
